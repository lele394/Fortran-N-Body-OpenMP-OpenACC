```python
import matplotlib.pyplot as plt
import numpy as np

import json

file = 'results.json'


# Load dictionary from a JSON file
with open(file, 'r') as json_file:
    times = json.load(json_file)

baseline_time = times['ST']
speedup = {key: baseline_time / value for key, value in times.items()}

labels = ["1 Thread", "23 Threads", "48 Threads", "GPU"]
values = list(speedup.values())

plt.bar(labels, values, color=['blue', 'green', 'cyan', 'red'])

for i, value in enumerate(values):
    plt.text(i, value + 0.2, f'{value:.2f}', ha='center', fontsize=10, fontweight='bold')

plt.xlabel('Configuration')
plt.ylabel('Speedup')
plt.title('Speedup of Different Configurations Relative to 1 Thread')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
